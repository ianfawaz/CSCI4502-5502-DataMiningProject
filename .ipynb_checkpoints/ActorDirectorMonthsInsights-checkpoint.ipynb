```python
actorMonths = pd.read_csv("actorMetaMonths.csv")
```

```python
actorMonths.head()
```

```python
actorMonths['revbudRatio'] = Series(float(0), index=actorMonths.index)
```

```python
for index, row in actorMonths.iterrows():
    #print(index)
    #print(type(row['movieIDs']))
#     if (type(row['movieIDs'])!=str):
#         continue
    budget = row['budget']
    revenue = row['revenue']
    #print(arrRow)
    actorMonths['revbudRatio'][index] = float(revenue/budget)
```

```python
actorMonths.head(50)
```

```python
actormonthSort = actorMonths.sort_values('revbudRatio', ascending=False)
actormonthSort.head(100)
```

```python
actormonthSort.shape
```

```python
januarymonth = actormonthSort[actormonthSort['january'] == 1]
```

```python
januarymonth['revbudRatio'].median()
```

```python
febmonth = actormonthSort[actormonthSort['february'] == 1]

febmonth['revbudRatio'].median()
```

```python
monthArr = ['january','february','march','april','may','june','july','august','september','october','november','december']
monthmeds = []
for i in range(len(monthArr)):
    month = actormonthSort[actormonthSort[monthArr[i]] == 1]
    monthmeds.append(month['revbudRatio'].median())
```

```python
print(monthmeds)
```

```python
    
sortArr = (sorted(range(len(monthmeds)),key=monthmeds.__getitem__))
sortArr = list(reversed(sortArr))
print(sortArr)
```

```python
for i in range(len(monthArr)):
    print(monthArr[sortArr[i]])
```
